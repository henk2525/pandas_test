###################################################
# 我們現在蒐集到了第二季的資料（fruit_sales_Q2.csv）#
###################################################

```python
revenue_Q2 = pd.read_csv('fruit_sales_Q2.csv')
revenue_Q2['Average'] = revenue_Q2[ ['Apple', 'Banana', 'Orange', 'Tomato', 'Guava'] ].mean( axis=1 )
```

# Q6：請將兩個資料集（revenue, revenue2）合併，並命名為 total

```python
total = pd.concat([revenue_Q1, revenue_Q2], ignore_index=True, axis=0)
total
```

# Q7：請確認 total 是否有合併完成

```python
print(total)
```

# Q8：請將 total 中的資料依照 Average 由高到低排序

```python
total = total.sort_values(by='Average', ascending=False)
total
```

# Q9：請將 total 輸出成 csv 檔案，名稱：自訂

```python
total.to_csv("winnie.csv")
```
